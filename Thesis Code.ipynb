```python
#Importing various libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree


```

```python
#Loading the Datasets
```

```python
train_data = pd.read_csv("train.csv")
ideal_data = pd.read_csv("ideal.csv")
test_data = pd.read_csv("test.csv")
```

```python
print(train_data)
```

```python
print(ideal_data)
```

```python
print(test_data)
```

```python
# Separating the features (x) and (y) in the training dataset
```

```python
train_x = train_data[['x']]
train_y = train_data[['y1']]
```

```python
print(train_x)
```

```python
print(train_y)
```

```python
# Train a linear regression model
linear_reg = LinearRegression()
linear_reg.fit(train_x, train_y)
```

```python
# Train a decision tree regression model
```

```python
tree_reg = DecisionTreeRegressor()
tree_reg.fit(train_x, train_y)
```

```python
# Using the mean squared error (MSE) to evaluate the models.
train_y_pred_linear = linear_reg.predict(train_x)
train_y_pred_tree = tree_reg.predict(train_x)
linear_mse = mean_squared_error(train_y, train_y_pred_linear)
tree_mse = mean_squared_error(train_y, train_y_pred_tree)
```

```python
# Print the models' MSEs.
print("Linear Regression MSE:", linear_mse)
print("Decision Tree Regression MSE:", tree_mse)
```

```python
# Visualize linear regression
plt.figure(figsize=(10, 6))
plt.scatter(train_x, train_y, color='blue', label='Actual')
plt.plot(train_x, train_y_pred_linear, color='red', linewidth=2, label='Linear Regression')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression')
plt.legend()
plt.show()

```

```python
# Visualize decision tree regression
plt.figure(figsize=(10, 6))
plot_tree(tree_reg, feature_names=['x'], filled=True)
plt.title('Decision Tree Regression')
plt.show()
```
